```python
X = data[['Symptom_1','Symptom_2','Symptom_3','Symptom_4','Symptom_5','Symptom_6']]
y = data['Disease']
```

```python
X.shape
```

```python
from sklearn.preprocessing import LabelEncoder

# Initialize LabelEncoder
label_encoder = LabelEncoder()

# Apply label encoding to each column
for col in X.columns:
    # Check if the column has missing values
    if X[col].isnull().any():
        # Create a mask to identify non-null values
        mask = X[col].notnull()
        # Apply label encoding only to non-null values
        X.loc[mask, col] = label_encoder.fit_transform(X.loc[mask, col])
    else:
        # Encode all values if no missing values
        X[col] = label_encoder.fit_transform(X[col])

y = label_encoder.fit_transform(y)

X
```

```python
from sklearn.impute import KNNImputer

# Select the columns for imputation
columns_to_impute = ['Symptom_4', 'Symptom_5', 'Symptom_6']

# Initialize KNNImputer
imputer = KNNImputer(n_neighbors=5)

# Perform KNN imputation for selected columns
X[columns_to_impute] = imputer.fit_transform(X[columns_to_impute])
```

```python
X.isna().sum()
```

```python
# Flatten all symptoms from all 6 columns
symptom_cols = ['Symptom_1', 'Symptom_2', 'Symptom_3', 'Symptom_4', 'Symptom_5', 'Symptom_6']
all_symptoms = pd.unique(data[symptom_cols].values.ravel())
all_symptoms = [s.strip().lower() for s in all_symptoms if pd.notnull(s)]
```

```python
# Build one-hot vectors for each row
def one_hot_encode_symptoms(row, all_symptoms):
    row_symptoms = [row[col].strip().lower() for col in symptom_cols if pd.notnull(row[col])]
    return [1 if symptom in row_symptoms else 0 for symptom in all_symptoms]

X_onehot = pd.DataFrame([one_hot_encode_symptoms(row, all_symptoms) for _, row in data.iterrows()], columns=all_symptoms)

```

```python
from sklearn.preprocessing import LabelEncoder
target_encoder = LabelEncoder()
y = target_encoder.fit_transform(data["Disease"])

```

```python
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_onehot, y, test_size=0.2, random_state=42)
model = XGBClassifier()
model.fit(X_train, y_train)
```

```python
def predict_disease(symptoms, model, all_symptoms, target_encoder):
    symptoms = [s.strip().lower() for s in symptoms]
    input_vector = [1 if symptom in symptoms else 0 for symptom in all_symptoms]
    input_df = pd.DataFrame([input_vector], columns=all_symptoms)
    pred = model.predict(input_df)
    return target_encoder.inverse_transform(pred)[0]
```

```python
user_input = ["patches_in_throat",
"high_fever",
"extra_marital_contacts"]
predicted_disease = predict_disease(user_input, model, all_symptoms, target_encoder)
print("ðŸ§  Predicted Disease:", predicted_disease)
```
